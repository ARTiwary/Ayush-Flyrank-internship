## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import os
import sys
import subprocess
import pandas as pd

# Setup for Google Colab
REPO_DIR = "flyrank-ml-internship-starter"

if "google.colab" in sys.modules:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(
            ["git", "clone", "--depth", "1",
             "https://github.com/flyrank-bih/flyrank-ml-internship-starter",
             REPO_DIR],
            check=True,
        )
    os.chdir(REPO_DIR)

# Load dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Create proxy target
df["is_declining"] = (df["trend_direction"] == "down").astype(int)

print("One row represents one content page.\n")

print(df[[
    "content_age_days",
    "days_since_last_update",
    "impressions_90d",
    "avg_position",
    "ctr",
    "trend_direction",
    "is_declining"
]].head())

print(f"\nTotal pages: {len(df)}")
print(f"Declining pages: {df['is_declining'].sum()}")
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
from IPython.display import Markdown, display

display(Markdown("""
### Why Machine Learning Instead of a Fixed Rule?

A simple rule such as "refresh every page older than 180 days" ignores many other signals that may influence page performance.

Features like impressions, click-through rate, average position, content age, and update history may interact in complex ways. Machine learning can learn patterns across multiple features simultaneously and produce a better ranking than a single manually written rule.

The model should be used as decision support rather than replacing human judgment.
"""))
```
