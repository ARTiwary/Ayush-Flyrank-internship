## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import sys
import subprocess
import pandas as pd

# Setup for Colab
REPO_DIR = "flyrank-ml-internship-starter"

if "google.colab" in sys.modules:
    if not os.path.isdir(REPO_DIR):
        subprocess.run(
            ["git", "clone", "--depth", "1",
             "https://github.com/flyrank-bih/flyrank-ml-internship-starter",
             REPO_DIR],
            check=True,
        )
    os.chdir(REPO_DIR)

# Load dataset
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print(f"Total pages in the dataset: {len(df)}")

corr_position_ctr = df["avg_position"].corr(df["ctr"])
print(f"Correlation between average position and CTR: {corr_position_ctr:.3f}")

corr_volume_impressions = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search volume and impressions: {corr_volume_impressions:.3f}")

print("\nMedian word count by trend direction:")
print(df.groupby("trend_direction")["word_count"].median())

print("\nObservations:")
print("- The dataset contains 30,000 pages, which is large enough to explore meaningful patterns.")
print("- Search volume has almost no correlation with impressions (0.001), suggesting that search volume alone is not enough to prioritize pages.")
print("- Average position and CTR show a negative relationship, indicating that ranking position influences click-through rate.")
print("- Median word counts are fairly similar across trend groups, suggesting that content length alone is unlikely to explain page performance.")
```
